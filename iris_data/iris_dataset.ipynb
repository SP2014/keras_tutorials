```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LogisticRegressionCV

from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.utils import np_utils
```

```python
# Load the dataset
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

```python
X = iris.values[:,:4]
Y = iris.values[:,4]

# Split the data into training and testing
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, train_size=0.5, random_state=0)
```

```python
# Train the model using Logistic Regression
lr = LogisticRegressionCV()
lr.fit(train_X, train_Y)
```

```python
print("Accuracy: {:.2f}".format(lr.score(test_X, test_Y)))
```

```python
# Keras specific

def one_hot_encode(arr):
    uniques,ids = np.unique(arr, return_inverse=True)
    return np_utils.to_categorical(ids, len(uniques))
```

```python
train_Y_ohe = one_hot_encode(train_Y)
test_Y_ohe = one_hot_encode(test_Y)

# Define the model
model = Sequential()
# Input layer
model.add(Dense(16, input_shape=(4,)))
model.add(Activation('sigmoid'))

# Output layer (as there are 3 classes)
model.add(Dense(3))
model.add(Activation('softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=["accuracy"])
```

```python
'''
Training neural networks often involves the concept of "minibatching", which means showing
the network a subset of the data, adjusting the weights, and then showing it another subset
of the data. When the network has seen all the data once, that's called an "epoch".
Tuning the minibatch/epoch strategy is a somewhat problem-specific issue, but in this case
we'll just use a minibatch of 1. That makes it effectively good old stochastic gradient descent,
i.e. the data is shown to the network one flower at a time, and the weights adjusted immediately.
'''
model.fit(train_X, train_Y_ohe, epochs=100, batch_size=2, verbose=0)
```

```python
loss, accuracy = model.evaluate(test_X, test_Y_ohe, verbose = 0)
print("Accuracy: {:.2f}".format(accuracy))
```
